```python
import numpy as np
from matplotlib import pyplot as plt
def sampling_lines_alt(ax, duration):

    x = np.linspace(0,130,70)
    y = x / duration
    ax.plot(x,y, label=f'∆ = {duration:.2f}')

    ax.set_xticks(np.arange(8, 100, 8.0))
    ax.set_yticks(np.arange(2, 25, 1.0))
    ax.set_xlim(8,100)
    ax.set_ylim(5,25)
    ax.set_xlabel('T')
    ax.set_ylabel('24 / τ')
    ax.grid(True)

    ax.legend()
    return
    ax.scatter([16, 32, 64], [6, 12, 24])
    x = np.linspace(0,130,70)
    y = x / (64 / 24)
    ax.plot(x,y, label=f'{(64 / 24):.2f}')

    ax.scatter([16, 32, 48, 96], [4, 8, 12, 24])
    x = np.linspace(0,130,70)
    y = x / (96 / 24)
    ax.plot(x,y, label=f'{(96 / 24):.2f}')

    ax.scatter([16, 32, 48, 64], [3, 6, 9, 12])
    x = np.linspace(0,130,70)
    y = x / (128 / 24)
    ax.plot(x,y, label=f'{(128 / 24):.2f}')

    ax.scatter([16, 32, 48, 64, 80], [2, 5, 7, 10, 12])
    x = np.linspace(0,130,70)
    y = x / (160 / 24)
    ax.plot(x,y, label=f'{(160 / 24):.2f}')

    return ax
```

```python
ax = sampling_lines_alt()
```

```python
fig, ax = plt.subplots(figsize=(10,10))

ax.set_title('R2+1D')

ax.scatter([64, 56, 48, 40, 32, 24, 16], [24, 21, 18, 15, 12, 9, 6])
sampling_lines_alt(ax, 64 / 24)
ax.arrow(32, 24, 32, 0, head_width=0.3, head_length=1, fc='k', ec='k', length_includes_head=True)
ax.annotate(60.6, (64, 24.1))
ax.arrow(32, 24, 16, -6, head_width=0.3, head_length=1, fc='k', ec='k', length_includes_head=True)
ax.annotate('?', (32, 12.1))
ax.arrow(32, 24, 0, -12, head_width=0.3, head_length=1, fc='k', ec='k', length_includes_head=True)
ax.annotate(59.6, (32, 12.1))

ax.scatter([96, 80, 72, 64, 56, 48, 40, 32, 24], [24, 20, 18, 16, 14, 12, 10, 8, 6])
sampling_lines_alt(ax, 96 / 24)
ax.arrow(32, 12, 16, 0, head_width=0.3, head_length=1, fc='k', ec='k', length_includes_head=True)
ax.annotate(61.4, (48, 12.1))

ax.scatter([96, 80, 64, 48, 32], [18, 15, 12, 9, 6])
sampling_lines_alt(ax, 128 / 24)

ax.scatter([32], [24])
sampling_lines_alt(ax, 32 / 24)
ax.annotate(58.2, (28, 24))

sampling_lines_alt(ax, 48 / 24)


fig.savefig('/Users/simonnarendorf/Documents/Projects/soccer-action-classification/Thesis/img/07_grid_r2plus1d.eps', format='eps')
```
